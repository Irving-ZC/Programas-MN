```python
#Primera función f(x)=x^2cos(x)-2x

def f(x):
       return x**2 * np.cos(x) - 2*x
def Df(x): #derivada de la funcion
    return 2*x*np.cos(x)-x**2*np.sin(x)-2

x0 = 1 #valor inicial
i=1 
error=10 #tolerancia
while error > 1e-6:

    x1= x0-f(x0)/Df(x0) #formula newton
    error= abs(x1-x0)
    x0= x1
    print("Iteración",i,", raíz aproximada:",x0)
    i=i+1

import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-2,2,100)
plt.plot(x,f(x))
plt.plot(x0,f(x0),'or') #indica la raiz
plt.grid()
plt.show()
```
